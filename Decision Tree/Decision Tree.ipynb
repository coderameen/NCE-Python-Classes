```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
# from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
#split into x and y axis
x = df.drop(columns=['salary_more_then_100k'])
x
```

```python
y = df.salary_more_then_100k
y
```

```python
#Data preprocessing: label encoding: Bcoz we have labes in my dataset
from sklearn.preprocessing import LabelEncoder
x['Company_l'] = LabelEncoder().fit_transform(x['company'])
x
```

```python
x = x.drop(columns=['company'])
x.head()
```

```python
#job
x['job_l'] = LabelEncoder().fit_transform(x['job'])
x = x.drop(columns=['job'])
#degree

x['degree_l'] = LabelEncoder().fit_transform(x['degree'])
x = x.drop(columns=['degree'])
x.head()
```

```python
#call the model
model  = DecisionTreeClassifier()
```

```python
#train the model
model.fit(x,y)
```

```python
#predict the outcomes
pred = model.predict([[1,]])
```
