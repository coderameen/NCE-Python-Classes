```python
import pandas as pd
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
dummy_gender = pd.get_dummies(df.Genre)
```

```python
dummy_gender
```

```python
df = pd.concat([df,dummy_gender],axis='columns')
df.head()
```

```python
df = df.drop(columns=['Genre'])
df.head()
```

```python
#call the model 
model  = DBSCAN(eps=3,min_samples=4)
```

```python
model.fit(df)
```

```python
model.fit_predict(df)
```
